```python
counter = 3036
image_array = np.array(Image.open('p04.jpg'))
print('original image shape (%d, %d)' % (image_array.shape[0], image_array.shape[1]))

for i in range(0, 23):
    increment = 0
    x_start = 4 + i * 51
    x_end = x_start + 48
    if i <= 5:
        y = 32
    else:
        y = 31
    for j in range(0, y):
        counter += 1
        if j < 12:
            y_start = 0 + j * 43
            y_end = y_start + 48
        if j >=12 and j < 23:
            increment += 1
            y_start = 0 + j * 43 + increment
            y_end = y_start + 48
        if j >= 23:
            y_start = 0 + j * 43 + increment
            y_end = y_start + 48
        image = image_array[y_start:y_end, x_start:x_end]
        im = Image.fromarray(image)
        im.save("SingleChar/%d.jpg" %counter)
print('counted:%d'%counter)
#     plt.subplot(8, 6, j)
#     plt.imshow(image, cmap='gray')
# plt.show()
```

```python
# invert the color of the grayscale images
in_dir = 'chars_orig/'
out_dir = 'chars_reverted/'
img_names = os.listdir(in_dir)
is_jpg = re.compile(r'.+?\.jpg')

# grayscale invert
for name in img_names:
    if(is_jpg.match(name)):
        image = Image.open(in_dir + name)
        inverted_image = ImageOps.invert(image)
        inverted_image.save(out_dir + name)
```

```python
# shift the image
img = np.array(Image.open('chars_reverted/1.jpg'))
row, col = img.shape
counter = 1

def imshow(img):
    plt.imshow(img, cmap='gray')
    plt.gca().axis('off')
    
plt.subplot(10, 10, counter)
counter += 1
imshow(img)

dst_list = []
for i in xrange(-3, 4):
    for j in xrange(-3, 4):
        plt.subplot(10, 10, counter)
        counter += 1
        M = np.float32([[1,0,3*i],[0,1,3*j]])
        dst = cv2.warpAffine(img, M, (col, row))
        imshow(dst)
        dst_list.append(dst)
    
plt.show()
```

```python
# rotate the image
angles = [-45, -30, -15, 15, 30, 45]
# angles = [-30]
counter = 1
for angle in angles:
    for i in xrange(0, 10):
        img_orig = dst_list[i]
        plt.subplot(10, 10, counter)
        counter += 1
        M = cv2.getRotationMatrix2D((col/2,row/2), angle, 1)
        dst = cv2.warpAffine(img_orig, M, (col,row))
        imshow(dst)

plt.show()
```

```python
# directMap
from scipy.misc import imread
import matplotlib.pyplot as plt
import numpy as np
import math

pi = math.pi
plt.rcParams['image.interpolation'] = 'none'

# show image helper
def imshow(img):
    plt.imshow(img, cmap='gray')
    plt.gca().axis('off')
    
# a naive conv forward    
def conv_forward_naive(x, w, b, conv_param):
    out = None
    H, W = x.shape
    HH, WW = w.shape
    pad = conv_param['pad']
    stride = conv_param['stride']

    x_padded = np.pad(x, ((pad, pad), (pad, pad)), 'constant')
    
    Out_h = (int)(1 + (H + 2 * pad - HH) / stride)
    Out_w = (int)(1 + (W + 2 * pad - WW) / stride)

    out = np.zeros((Out_h, Out_w))

    for k in range(Out_h):
        for l in range(Out_w):
            out[k, l] = np.sum(
                x_padded[k * stride : k * stride + HH, l * stride : l * stride + WW] * w) + b
            
    return out

# sobel operation using conv            
def sobel(x, w_x, w_y):
    g_x = conv_forward_naive(x, w_x, b, {'stride': 1, 'pad': 1})
    g_y = conv_forward_naive(x, w_y, b, {'stride': 1, 'pad': 1})
    g_abs = np.abs(g_x) + np.abs(g_y) 
    # use arctan2 to compute [-pi, pi] angles
    g_arc = np.arctan2(g_y, g_x)
    return (g_abs, g_arc)

# map the gradients into directMap
def grad2directMap(g_abs, g_arc):
    directMap = np.zeros((8, g_abs.shape[0], g_abs.shape[1]))
    arcs = np.array([0, pi/4, pi/2, pi*3/4, pi])
    for i in range(4):
        lbound = arcs[i]
        ubound = arcs[i+1]
        mask = (g_arc < ubound) * (g_arc >= lbound) 
        directMap[i, :, :] += mask * g_abs * (np.cos(g_arc - lbound) - np.sin(g_arc - lbound))
        directMap[i+1, :, :] += mask * g_abs * (np.cos(ubound - g_arc) - np.sin(ubound - g_arc))
    arcs = np.array([-pi, -pi*3/4, -pi/2, -pi/4, 0])
    for i in range(4):
        lbound = arcs[i]
        ubound = arcs[i+1]
        mask = (g_arc < ubound) * (g_arc >= lbound) 
        directMap[i+4, :, :] += mask * g_abs * (np.cos(g_arc - lbound) - np.sin(g_arc - lbound))
        if(i == 3):
            directMap[0, :, :] += mask * g_abs * (np.cos(ubound - g_arc) - np.sin(ubound - g_arc))
        else:
            directMap[i+5, :, :] += mask * g_abs * (np.cos(ubound - g_arc) - np.sin(ubound - g_arc))
    directMap[directMap>255] = 255
    return directMap        
        
# sobel operator
w_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
w_x = np.array([[1, 0, -1],[2, 0, -2],[1, 0, -1]])
b = 0
```

```python
img_0 = imread('chars_reverted/2.jpg')

# perform sobel and get adsolute value and direction of the input image
img_g_abs_0, img_g_arc_0 = sobel(img_0, w_x, w_y)
directMap_0 = grad2directMap(img_g_abs_0, img_g_arc_0)

# show the outcome
plt.subplot(1, 10, 1)
# plt.title('original')
imshow(img_0)
for i in range(2, 10):
    plt.subplot(1, 10, i)
    kernel = np.ones((2,2),np.uint8)
    directMap_0[i-2, :, :] = cv2.erode(directMap_0[i-2, :, :], kernel, iterations=1)
    imshow(directMap_0[i-2, :, :])

directMap_added = np.sum(directMap_0, axis=0)
directMap_added[directMap_added>255] = 255 
plt.subplot(1, 10, 10)
imshow(directMap_added)
Image.fromarray(directMap_added).convert('L').save("directMap.jpg")

plt.show()
```
